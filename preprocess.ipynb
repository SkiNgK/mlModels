```python
import pyedflib
import pandas as pd
import numpy as np
import os
```

```python
raiz = r'''C:\Users\jonat\OneDrive\Documents\Protocolo_PKS\Protocolo_PKS\dados_edf\parkinson\1\Francisco Alc√¢ntara - Dia 17.11.08 as 14.53.32.edf'''
```

```python
edf = pyedflib.EdfReader(raiz)
```

```python
edf.getNSamples()
```

```python
n = edf.signals_in_file
```

```python
signal_labels = edf.getSignalLabels()
```

```python
sigbufs = np.zeros((n, edf.getNSamples()[0]))
```

```python
sigbufs
```

```python
for i in np.arange(n):
     sigbufs[i, :] = edf.readSignal(i)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
labels =['Ch 1', 'Ch 2', 'Ch 3', 'Ch 4']
```


Each file contains 5 columns, in this way:

Segment 	Leg

Channel	Ch1	Ch2	Ch3	Ch4

Muscle	RF	BF	VM	ST

Column	0	1	2	3

* RF: rectus femoris.
* BF: biceps femoris.
* VM: vastus internus.
* ST: semitendinosus.

```python
allFeatures = pd.DataFrame(sigbufs.T, columns=labels)
```
